```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/partA.csv')
```

```python
df
```

```python
yerr = [[0.04],[0.04]]
plt.errorbar(df['d'],df['T'], yerr=yerr, capsize=3, ecolor = "black", ls='none')
plt.scatter(df['d'],df['T'],s=8,label='data points')
plt.plot(df['d'][:11],df['T'][:11],color='blue')
plt.plot(df['d'][11:],df['T'][11:],color='blue')
plt.xlabel('distance from center (cm)')
plt.ylabel('time period of pendulum (s)')
plt.axhline(1.68,color='red',label='t1*')
plt.axhline(1.71,color='orange',label='t2*')
plt.axhline(1.74,color='green',label='t3*')
plt.legend()
plt.ylim(1.5,2)
plt.show()
```
